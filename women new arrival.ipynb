```python
S_color_lst =['Black','Blue','Brown','Cyan','Golden','Green','Grey','Maroon','Olive','Orange','Pink','Purple','Red','White','Yellow','Magenta']
S_febric_lst = ['Addi Cotton','Cotton','Endi Cotton','Endi Silk','Joysree Silk','Mixed Cotton','Mixed Silk','Mixed Viscose','Ramie Cotton','Sarin','Silk','Viscose','Mixed Muslim','','','']
find_types = []
item_categorys = []
name_lst = []
list_id = []
image_source = []
prices = []
color_lst = []
febric_lst = []
counter = 0
for i in range(0,54):
    url=f'https://www.aarong.com/women/women-new-arrivals?_=1670752886042&p={i}'
    print(url)
    # Make a GET request to fetch the raw HTML content
    html_content = requests.get(url).text
    # Parse the html content
    soup = BeautifulSoup(html_content, "lxml")
    order_list = soup.find_all("li", class_="item product product-item")
    find_category = soup.find("div", class_="breadcrumbs")
    item_category = find_category.find("strong").text
    find_type1 = soup.find("li", class_="item category4")
    item_type1 = find_type1.find("a").text.replace('\n ','')

    for li in order_list:
        #get item type
        find_types.append(item_type1.replace(' ',''))
        #get iteam category
        item_categorys.append(item_category)
        
        ## get element name
        w_not = False
        f_not =False
        title_element = li.find("a", class_="product-item-link").text.replace('\n ','')
        chek=0
        chek2 = 0
        for word in title_element.split():
            
            if word in S_color_lst:
                w_not = True
                chek+=1
                if chek == 1:
                    color_lst.append(word)
                if chek==2:
                    print(" ".join(title_element.split()),chek)
            if word in S_febric_lst:
                f_not =True
                chek2 += 1
                if chek2==1:
                    febric_lst.append(word)
            
#         print(" ".join(title_element.split()),counter,w_not,f_not)
        if w_not is False:
            color_lst.append('No color specified')
        if f_not is False:
            febric_lst.append('No febric specified')
        name_lst.append(" ".join(title_element.split()))
        #get picture link
        link = li.find("img", class_="product-image-photo").get('src')
        image_source.append(link)
        # get price
        price = li.find("span", class_="price-wrapper").get("data-price-amount")
        prices.append(float(price))
        ### set id
        list_id.append(counter)
        counter +=1
       # print(len(febric_lst))
```

```python
len(find_types),len(item_categorys),len(color_lst),len(febric_lst),len(name_lst),len(image_source),len(prices),len(list_id)
```

```python
data = {'item type':find_types,'item category':item_categorys,'color':color_lst,'febric':febric_lst,'name':name_lst,'image':image_source,'price':prices,'product classification id':list_id}
data
```

```python
df = pd.DataFrame(data=data)
df
```

```python
df.to_csv('arong_women_new_arrivals.csv')
```
