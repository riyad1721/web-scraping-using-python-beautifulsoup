```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.aarong.com/men/men-new-arrivals'
html_request = requests.get(url).text
html_request
```

```python
page_url_lst = []
for i in range(0,74):
    url=f'https://www.aarong.com/men/men-new-arrivals?_=1670489565784&p={i}'
    # Make a GET request to fetch the raw HTML content
    html_content = requests.get(url).text
    # Parse the html content
    soup = BeautifulSoup(html_content, "lxml")
    order_list = soup.find_all("li", class_="item product product-item")
    for li in order_list:
        lst_page_url = li.find("a", class_="product-item-link").get('href')
        page_url_lst.append(lst_page_url)
        
```

```python
len(page_url_lst)
```

```python
data ={'page url link': page_url_lst}
data
```

```python
df = pd.DataFrame(data=data)
df
```

```python
df.to_csv('page_url_link.csv')
```
