```python
iteam_for = []
item_type = []
item_fabric = []
item_name = []
image_one = []
image_two = []
image_three = []
image_four = []
item_prize = []
avilable_size = []
product_code = []
Collar_or_Neck = []
Cut_or_Fit = []
specification_Fabric = []
specification_Pocket = []
specification_Side_Cut = []
specification_Length = []
specification_Value_Addition = []
```

```python
for i in range(888):
    request = requests.get(data_link['page url link'][i],headers)
    new_soup = BeautifulSoup(request.content, 'html.parser')
    item_for_tag = new_soup.find('li', class_='item 1')
    if item_for_tag is not None:
        iteam_for.append(item_for_tag.find('span').text)
    else:
        iteam_for.append('not specified')
    item_type_tag = new_soup.find('li', class_='item 2')
    if item_type_tag is not None:
        item_type.append(item_type_tag.find('span').text)
    else:
        item_type.append('not specified')
    item_fabric_tag = new_soup.find('li', class_='item 3')
    if item_fabric_tag is not None:
        item_fabric.append(item_fabric_tag.find('span').text.replace('amp;',''))
    else:
        item_fabric.append('not specified')
    item_name_tag = new_soup.find('h1', class_='page-title')
    item_name.append(item_name_tag.find('span').text)
    product_image_collect_tag = new_soup.find("div",class_='column main')
    get_script =  product_image_collect_tag.find_all('script',type="text/x-magento-init")
    if i == 335 or i == 392:
        get_script = get_script[4]
    else:
        get_script = get_script[7]
    json_object = json.loads(get_script.contents[0])
    img_status_1 = 0
    img_status_2 = 0
    img_status_3 = 0
    img_status_4 = 0
    for j in range(len(json_object['[data-gallery-role=gallery-placeholder]']['mage/gallery/gallery']['data'])):
        if j == 0:
            image_one.append(json_object['[data-gallery-role=gallery-placeholder]']['mage/gallery/gallery']['data'][0]['full'])
            img_status_1 += 1
        elif j == 1:
            image_two.append(json_object['[data-gallery-role=gallery-placeholder]']['mage/gallery/gallery']['data'][1]['full'])
            img_status_2 += 1
        elif j == 2:
            image_three.append(json_object['[data-gallery-role=gallery-placeholder]']['mage/gallery/gallery']['data'][2]['full'])
            img_status_3 += 1
        elif j == 3:
            image_four.append(json_object['[data-gallery-role=gallery-placeholder]']['mage/gallery/gallery']['data'][3]['full'])
            img_status_4 += 1
    if img_status_1 == 0:
         image_one.append('not specified')
    if img_status_2 == 0:
         image_two.append('not specified')
    if img_status_3 == 0:
         image_three.append('not specified')
    if img_status_4 == 0:
         image_four.append('not specified')
    price_tag =new_soup.find('div',class_='product-info-price')
    if price_tag.find('span',class_='price-wrapper') is not None:
        item_prize.append(new_soup.find('span',class_='price-wrapper').get('data-price-amount'))
    else:
        item_prize.append(price_tag.find('span').text.replace('\n',''))
    size_tag = new_soup.find('div',class_='product-options-wrapper')
    if size_tag is not None:
        size_script = size_tag.find('script')
        size_json = json.loads(size_script.contents[0])
        d_tuple = ()
        if i == 887:
            print('calling',i)
            for k in range(len(size_json['#product_addtocart_form']['configurable']['spConfig']['attributes']['376']['options'])):
                d_tuple = d_tuple + (size_json['#product_addtocart_form']['configurable']['spConfig']['attributes']['376']['options'][k]['label'],)
        else:
            for k in range(len(size_json['#product_addtocart_form']['configurable']['spConfig']['attributes']['794']['options'])):
                d_tuple = d_tuple + (size_json['#product_addtocart_form']['configurable']['spConfig']['attributes']['794']['options'][k]['label'],)
        avilable_size.append(d_tuple)
    else:
        avilable_size.append((0,0,0))
    product_code_tag = new_soup.find('div',class_='product attribute Sku')
    product_code.append(product_code_tag.find('div',class_="value").text)
    product_specification_tag = new_soup.find('div',class_='additional-attributes-wrapper table-wrapper')
    product_specification_th_tag = product_specification_tag.find_all('th',class_='col label')
    product_specification_td_tag = product_specification_tag.find_all('td',class_='col data')
    status_1 = 0
    status_2 = 0
    status_3 = 0
    status_4 = 0
    status_5 = 0
    status_6 = 0
    status_7 = 0
    for th,td in zip(product_specification_th_tag,product_specification_td_tag):
        if th.text == 'Collar/Neck':
            Collar_or_Neck.append(td.text)
            status_1 +=1
        elif th.text == 'Cut /Fit':
            Cut_or_Fit.append(td.text)
            status_2 +=1
        elif th.text == 'Fabric':
            specification_Fabric.append(td.text)
            status_3 +=1
        elif th.text == 'Pocket':
            specification_Pocket.append(td.text)
            status_4 +=1
        elif th.text == 'Side Cut':
            specification_Side_Cut.append(td.text)
            status_5 +=1
        elif th.text == 'Length' and status_6 == 0:
            specification_Length.append(td.text)
            status_6 +=1
        elif th.text == 'Value Addition':
            specification_Value_Addition.append(td.text)
            status_7 +=1
    if status_1 == 0 :
        Collar_or_Neck.append('not specified')
    if status_2 == 0:
        Cut_or_Fit.append('not specified')
    if status_3 == 0:
        specification_Fabric.append('not specified')
    if status_4 == 0:
        specification_Pocket.append('not specified')
    if status_5 == 0:
        specification_Side_Cut.append('not specified')
    if status_6 == 0:
        specification_Length.append('not specified')
    if status_7 == 0:
        specification_Value_Addition.append('not specified')
```
